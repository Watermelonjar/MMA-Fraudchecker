```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
r = requests.get('https://www.tapology.com/rankings/current-top-ten-best-pound-for-pound-mma-and-ufc-fighters')
```

```python
r.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
a_tags = soup.find_all('a', title=True,)

texts = [a['title'] for a in a_tags]
for text in texts:
  if 'items' in text:
    texts.remove(text)
texts
```

```python
class Fighter:
    def __init__(self, name, reach, stance, height, weight, wins, losses, draws):
        self.name = name
        self.reach = reach
        self.stance = stance
        self.height = height
        self.weight = weight
        self.wins = wins
        self.losses = losses
        self.draws = draws

    def win_rate(self):
        if self.wins + self.losses == 0:
            return 0
        return self.wins / (self.wins + self.losses)

    def __str__(self):
      return f"Fighter Name: {self.name}\nReach: {self.reach}\nStance: {self.stance}\nHeight: {self.height}\nWeight: {self.weight}\nWins: {self.wins}\nLosses: {self.losses}\nDraws: {self.draws}\nWin Rate: {self.win_rate():.2f}"

jon_jones = Fighter("Jon Jones", 84, "Orthodox", 78, 205, 26, 1, 0)
print(jon_jones)
```
